#### - Velocidade: 
   - numpy é até 10x mais rápido do que python's list.
   - array tem tamnho fixo (Numpy arrays are fixed in size) ao contrário de list que podem mudar de tamanho.
   - arrays por serem de tamanho único conseguem otimizar melhor a memória
   
   
#### -Funcionalidades
   - há várias funcionalidades para manipulação de matrizes (arrays multidimensionais)

### Definindo os tipos de um array

```python
np.array([], bool)
```

```python
array_simples = np.array([1, 33, 103, 1])
print(array_simples[0], array_simples[2])
print ('Tamanho do array eh: {}' .format(len(array_simples)))
print(type(array_simples))
print(array_simples.dtype) 

# mix types
array_misto = np.array(['zero', 323])
print(array_misto)
print(type(array_misto))
print(array_misto.dtype) 

# formato de uma matriz
print(array_simples.shape)
print(array_misto.shape)
```

### Funções de arrays

```python
print('Máximo valor: ', array_simples.max())
print('Valor médio: ', array_simples.mean())
print('Desvio padrão: ', array_simples.std())
print('Variância: ', array_simples.var())
print('Soma: ', array_simples.sum())
```

### Criação de arrays

```python
# Veja que eh uma funcionalidade da biblioteca numPy
a = np.zeros((2,2))   # Create an array of all zeros
print(a)              # Prints "[[ 0.  0.]
                      #          [ 0.  0.]]"
```

```python
b = np.ones((1,2))    # Create an array of all ones
print(b)              # Prints "[[ 1.  1.]]"
```

```python
np.identity(8, dtype=int)
```

```python
c = np.full((2,2), 7)  # Create a constant array
print(c)               # Prints "[[ 7.  7.]
                       #          [ 7.  7.]]"
```

## Operações entre entre matrizes

Aqui vamos começar a trabalhar com conceitos de eixos. Em python o eixo 0 representa as linhas de uma matriz, enquanto o eixo 1 representa as colunas.

```python
a = np.array([[1, 2], [3, 4]], float)
b = np.array([[5, 6], [7,8]], float)
```

```python
np.concatenate((a,b))
```

## Vamos falar sobre matemática
Um dos grandes problemas de trabalhar com listas ao invés de vetores, é executar operações matemáticas em todos os elementos da lista. Em geral, em uma lista, precisamos percorrer todos os elementos para fazer as operações. 

Operações matemáticas com vetores e matrizes são muito mais rápidas e eficientes do que iterações. O Numpy permite que esse tipo de operação aconteça de forma muito simples:

```python
a = np.array([1,2,3], float)
b = np.array([5,2,6], float)
```

```python
a + b
```

```python
a - b
```

```python
a * b
```

```python
b / a
```

```python
a % b
```

```python
b ** a
```

```python
a > b
```

Arrays com tamanhso diferentes:

```python
# o array de maior dimensão eh que define a operação
a = np.array([[0,0],[0,0]])
b1 = np.array([1,1])
b2 = 1
```

```python
a+b1
```

```python
b2+b1
```
